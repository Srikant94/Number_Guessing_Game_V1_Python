```python
import random

num = random.randrange(1,20) 

user_input = int(input('Enter the number'))

while user_input != num :
    
    if user_input < num :
        print('number is small')
        user_input = int(input('Enter the number again'))
    
    elif user_input > num :
        print('number is large')
        user_input = int(input('Enter the number again'))
    else :
        break

print('number is correct')
```
