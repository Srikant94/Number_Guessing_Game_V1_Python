```python
# Version 2 - Number Guessing Game with difficulty levels
# This version introduces multiple difficulty levels, dynamic attempt limits
# Improved scoring logic and replay functionality 

print('Welcome to Number Guessing Game')

# Importing Random module to generate random numbers

import random

# Function to allow the user to select different difficulty level 
# Each difficulty level defines its own number range and maximum attempts

def difficulty_level () :
    print('Please select the difficulty level :- ') 
    print('1. Easy') # -> Range - 1-10 ; Attempts - 7')
    print('2. Medium') # -> Range - 1-20 ; Attempts - 5')
    print('3. Hard') # -> Range - 1-30 : Attempts - 3')

# Loop until the user provides a valid difficulty level

    while True :
        level = input('Select - 1 / 2 / 3')
        print(f'Level Selected - {level}')
  
        if level == '1' :
            return 10, 7
        elif level == '2' :
            return 20, 5
        elif level == '3' :
            return 30, 3
        else :
            print('Invalid Choice')

# Core game function that controls the game flow and guessing logic

def number_game () :

# Retrieving number range and attempt limit based on difficulty level selected by user

    maximum_range, maximum_attempts = difficulty_level()
 
# Generating random number within the selected number range

    random_number = random.randint(1,maximum_range) 

# Initialising attempts and score

    attempts_left = maximum_attempts
    score = 0

# Displaying game instructions to the user  

    print(f'You have to guess the number between 1 to {maximum_range}') 
    print(f'You have total {maximum_attempts} attempts')

# Loop continues until the user runs out of attempt limit

    while attempts_left > 0 :
        try :

# Accepting and converting user input to integar
            
            user_inp = int(input('Enter your number : '))
        except ValueError :

# Handling non-numeric input gracefully

            print('Please provide a valid number')
            continue 

# Checking if the user guessed input is correct and matches with the random number

        if user_inp == random_number :
            score = attempts_left * 15
            print(f'Number entered - {user_inp}')
            print(f'You guessed the correct number - {random_number}')
            print(f'Score : {score}')
            break

# Providing hint if the guess is high

        elif user_inp > random_number :
            attempts_left -= 1
            print(f'Number entered - {user_inp}')
            print('Number is large')
            
        else :

# Providing hint if the guess is low

            attempts_left -= 1
            print(f'Number entered - {user_inp}')
            print('Number is small')

# Displaying remaining attempts left after each guess        
        
        print(f'Attempts Left : {attempts_left}')

# Informing the user once all the attempts are exhausted

    if attempts_left == 0 :
        print('No further attempts left now! Game Over')
        print(f'Score : {score}')
        print(f'The correct number is : {random_number}')    

# Defining the main function to enable replay functionality

def main ():
    while True :
        number_game()
        replay = input('Do you want to replay the game - Yes/No ').lower()

        if replay != 'yes':
            print('Thanks for playing')
            break
# Entry point of the game

main()    
```
